```python
import torch
from transformers import pipeline

model_id = "meta-llama/Llama-3.2-1B-Instruct"
pipe = pipeline(
    "text-generation",
    model=model_id,
    dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
system = """
You are and assistant and you are going to be given a list of words separed by ' '.
Your main objective is from the given the unordered words deduce the book and author and give
the response should only by in JSON format. 

# Steps
1. Identify the different words given. Important: you should only use the words given.
2. Guess the book name title from the given words
3. Guess the author of the book from the given words

# Response example
{
'author': author name,
'book title': book title
}
"""

book_words = "the taleb . 2 2 to black swain nassim . nicholas"

messages = [
    {"role": "system", "content": system},
    {"role": "user", "content": book_words},
]

outputs = pipe(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1]['content'])
```
