## YOLO

First we will need to load the model

```python
# This code snippet will look at the image and get the objects
from ultralytics import YOLO

# Load a pretrained YOLO26n model
model = YOLO("../models/yolo26n.pt")

img = Image.open("../data/OD/IMG_20260116_211626113.jpg")

results = model(img, conf=0.3)

# Draw bounding boxes
# for result in results:
#     boxes = result.boxes
#     detected = boxes.cls
#     bounding = boxes.xyxy
#     for bx,cls in zip(bounding,detected):
    #     draw = ImageDraw.Draw(img)
    #     x1, y1, x2, y2 = bx
    #     draw.rectangle([x1, y1, x2, y2], outline="red", width=3)
    # img.show()

# Crop image and save the rectangle
for result in results:
    boxes = result.boxes
    detected = boxes.cls
    bounding = boxes.xyxy
    for i, bx in enumerate(bounding):
        x1, y1, x2, y2 = bx.tolist()
        cropped = img.crop((x1, y1, x2, y2))
        cropped.save(f"crops/bbox_{i}.jpg")
```
